```python
#counting how many samples we have of both groups.
c_sam = [c for c in df.columns if c.endswith('_C')]
d_sam = [c for c in df.columns if c.endswith('_D')]

len(c_sam), len(d_sam)
```

```python
#getting the average of both groups in the data.
df['c_mean'] = df[c_sam].mean(axis=1)
df['d_mean'] = df[d_sam].mean(axis=1)
```
